# Модель

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import RobustScaler
```

```python
X = df_train.drop(['blue', 'clock_speed', 'm_dep', 'n_cores', 'four_g', 'wifi', 'dual_sim','price_range'], axis=1)
target = df_train['price_range']
```

```python
scaler = RobustScaler()
X_train = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.25, random_state=7, shuffle=True)
rnf = RandomForestClassifier()
rnf.fit(X_train, y_train)
y_pred = rnf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

Получаем модель с точностью более 87%. Теперь используя его определим ценовую категорию телефонов из нужного датасета.
